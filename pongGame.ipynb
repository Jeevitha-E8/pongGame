```python
import turtle
wd=turtle.Screen()
wd.title("Pong Game by Jeevitha")
wd.setup(width=500,height=400)
wd.bgcolor("black")
wd.tracer(0)

# Creating Paddle_A
paddle_A=turtle.Turtle()
paddle_A.speed(0)
paddle_A.shape("square")
paddle_A.shapesize(stretch_wid=5,stretch_len=1)
paddle_A.color("white")
paddle_A.penup()
paddle_A.goto(-200,0)

# Creating Paddle_B
paddle_B=turtle.Turtle()
paddle_B.speed(0)
paddle_B.shape("square")
paddle_B.shapesize(stretch_wid=5,stretch_len=1)
paddle_B.color("white")
paddle_B.penup()
paddle_B.goto(200,0)

# Creating Ball

ball=turtle.Turtle()
ball.shape("circle")
ball.color("white")
ball.penup()
ball.goto(0,0)
ball.speed(0)
ball.dx=0.1
ball.dy=0.1

#Creating pen

pen=turtle.Turtle()
pen.speed(0)
pen.penup()
pen.color("white")
pen.hideturtle()
pen.goto(0,160)
pen.write("Player A: 0 Player B: 0",align="center",font=("Courier",15,"normal"))

paddle_a_pen=0
paddle_b_pen=0
# Move Paddle_A_Up
def Paddle_A_up():
    y=paddle_A.ycor()
    y += 20
    paddle_A.sety(y)
wd.listen()
wd.onkeypress(Paddle_A_up,"Right")

# Move Paddle_A_down
def Paddle_A_down():
    y=paddle_A.ycor()
    y -= 20
    paddle_A.sety(y)
wd.listen()
wd.onkeypress(Paddle_A_down,"Left")

# Move Paddle_B_up
def Paddle_B_up():
    y=paddle_B.ycor()
    y += 20
    paddle_B.sety(y)
wd.listen()
wd.onkeypress(Paddle_B_up,"Up")

# Move Paddle_B_down
def Paddle_B_down():
    y=paddle_B.ycor()
    y -= 20
    paddle_B.sety(y)
wd.listen()
wd.onkeypress(Paddle_B_down,"Down")

# Main class
while True:
    wd.update()
    
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    
    # Border Checking
    if ball.ycor()>190:
        ball.sety(190)
        ball.dy *= -1
    if ball.ycor()<-190:
        ball.sety(-190)
        ball.dy *= -1
    if ball.xcor()>240:
        paddle_a_pen+=1
        pen.clear()
        ball.goto(0,0)
        ball.dx *= -1
        pen.write("Player A: {} Player B: {}".format(paddle_a_pen,paddle_b_pen),align="center",font=("Courier",15,"normal"))
    if ball.xcor()<-240:
        ball.goto(0,0)
        pen.clear()
        paddle_b_pen+=1
        ball.dx *= -1
        pen.write("Player A: {} Player B: {}".format(paddle_a_pen,paddle_b_pen),align="center",font=("Courier",15,"normal"))
    # Ball touches paddles
    if ball.xcor()<-190  and ball.ycor()<paddle_A.ycor()+50 and ball.ycor()>paddle_A.ycor()-50:
        ball.setx(-190)
        ball.dx *= -1
    if ball.xcor()>190  and ball.ycor()<paddle_B.ycor()+50 and ball.ycor()>paddle_B.ycor()-50:
        ball.setx(190)
        ball.dx *= -1
```
